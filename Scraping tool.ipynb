```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import requests
from PIL import Image
from selenium.webdriver.common.keys import Keys
from io import StringIO
```

```python
url = 'https://www.littlegiger.com/gigerdb/index.php'
```

```python
wd = webdriver.Chrome(executable_path='./chromedriver.exe')
wd.get(url)
```

```python
no_of_jobs = 703
```

```python
find_element = '/html/body/table/tbody/tr[2]/td[2]/center/table/tbody/tr/td/center/form/table/tbody/tr[2]/td[2]/table/tbody/tr/td[2]/input' 
search_element = '/html/body/table/tbody/tr[2]/td[2]/center/table/tbody/tr/td/center/form/table/tbody/tr[2]/td[2]/table/tbody/tr/td[2]/input'
```

```python
def save_image(i,url):
    image_name =  "Images//"+str(i)+'.jpg'
    data = requests.get(url).content
    with open (image_name, 'wb') as f:
        f.write(data)
```

```python
i = 1
while i < no_of_jobs:
    searcher = wd.find_element_by_name("SearchText")
    
    searcher.send_keys(str(i))
    time.sleep(2)
    
    #wd.find_element_by_class_name('StartButton').click()
    searcher.send_keys(Keys.ENTER)
    searcher.clear()
    time.sleep(3)
    try:
        img = wd.find_element_by_class_name('PrevImg').get_attribute("src")
        print(img)
        save_image(i,img)
    except:
        print("Could not find image" + str(i))
    i += 1
```
